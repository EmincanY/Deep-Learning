```python
X = df.drop("Type" , axis=1).values
y = df["Type"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size= 0.3 , random_state=25)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation , Dropout

from tensorflow.keras.callbacks import EarlyStopping
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(units = 30 , activation = "relu"))
model.add(Dense(units = 15 , activation = "relu"))
model.add(Dense(units = 15 , activation = "relu"))

model.add(Dense(units = 1 , activation = "sigmoid")) # Classfication output layer need activation. And you can use sigmoid activation function in there.

model.compile(loss = "binary_crossentropy" , optimizer = "adam")
```

```python
model.fit(X_train , y_train , validation_data=(X_test,y_test) , verbose = 1 , epochs= 700)
```

```python
model.history.history
```

```python
modelLoss = pd.DataFrame(model.history.history)
```

```python
modelLoss.plot(); # This model have overfitting.
```
